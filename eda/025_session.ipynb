```python
import pandas as pd
import numpy as np
import tqdm
import warnings
import matplotlib.pyplot as plt

pd.set_option("max_rows", 200)
pd.set_option("max_columns", 50)
tqdm.tqdm.pandas()
%matplotlib inline
```

```python
df = pd.read_pickle("../input/riiid-test-answer-prediction/split10/train_0.pickle")
```

```python
import sys
sys.path.append("../")
# df = pd.read_pickle("../input/riiid-test-answer-prediction/split10/train_0.pickle")
df = df.sort_values(["user_id", "timestamp"])
# large_user_id = df["user_id"].value_counts()
# large_user_id = large_user_id[large_user_id > 1000] 
# df = df[df["user_id"].isin(large_user_id.index)]
# df = df[df["answered_correctly"] != -1]
```

```python
df["answered_correctly"] = df["answered_correctly"].replace(-1, np.nan)
```

```python
df["timestampdiff"] = df["timestamp"] - df.groupby("user_id")["timestamp"].shift(1)
```

```python
df["change_session"] = (df["timestampdiff"] > 1000000).astype("int8")
```

```python
df["session"] = df.groupby("user_id")["change_session"].cumsum()
```

```python
df["session_nth"] = df.groupby(["user_id", "session"]).cumcount()
```
