```python
import pandas as pd
import numpy as np
import tqdm
import warnings
import matplotlib.pyplot as plt

pd.set_option("max_rows", 200)
pd.set_option("max_columns", 50)
tqdm.tqdm.pandas()
%matplotlib inline
```

```python
df = pd.read_pickle("../input/riiid-test-answer-prediction/split10/train_0.pickle")
```

```python
import sys
sys.path.append("../")
# df = pd.read_pickle("../input/riiid-test-answer-prediction/split10/train_0.pickle")
df = df.sort_values(["user_id", "timestamp"])
# large_user_id = df["user_id"].value_counts()
# large_user_id = large_user_id[large_user_id > 1000] 
# df = df[df["user_id"].isin(large_user_id.index)]
# df = df[df["answered_correctly"] != -1]
```

```python
df["is_bundled"] = \
(df.groupby("user_id")["bundle_id"].shift(-1) == df["bundle_id"]).astype("int8") + \
(df.groupby("user_id")["bundle_id"].shift(1) == df["bundle_id"]).astype("int8")
df["is_bundled"] = (df["is_bundled"] > 0)
```

```python
df = df[df["is_bundled"] == False]
```

```python
df["answered_correctly"] = df["answered_correctly"].replace(-1, np.nan)
```

```python
df["shift1"] = df.groupby("user_id")["answered_correctly"].shift(1).fillna(9).astype(int).astype(str)
df["shift2"] = df.groupby("user_id")["answered_correctly"].shift(2).fillna(9).astype(int).astype(str)
df["shift3"] = df.groupby("user_id")["answered_correctly"].shift(3).fillna(9).astype(int).astype(str)
```

```python
df["prev_ans"] = df["shift1"] + df["shift2"] + df["shift3"]
```

```python
df["user_mean"] = df.groupby("user_id")["answered_correctly"].transform("mean")
```
