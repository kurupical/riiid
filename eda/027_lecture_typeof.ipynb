```python
import sys
sys.path.append("../")
# df = pd.read_pickle("../input/riiid-test-answer-prediction/split10/train_0.pickle")
df = df.sort_values(["user_id", "timestamp"])
# large_user_id = df["user_id"].value_counts()
# large_user_id = large_user_id[large_user_id > 1000] 
# df = df[df["user_id"].isin(large_user_id.index)]
# df = df[df["answered_correctly"] != -1]
```

```python
df["answered_correctly"] = df["answered_correctly"].replace(-1, np.nan)
```

```python
df["lecture_count"] = df.groupby("user_id")["content_type_id"].cumsum()
```

```python
df.groupby("lecture_count")["answered_correctly"].describe().head(100)
```

```python
for col in ["concept", "solving question", "intention"]:
    df[col] = df["type_of"] == col
    df[f"{col}_cnt"] = df.groupby("user_id")[col].cumsum()
```

```python
df[df["timestamp"] > 10**6].groupby("concept_cnt")["answered_correctly"].describe()
```
