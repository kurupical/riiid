```python
import pandas as pd
import numpy as np
import tqdm
import warnings

pd.set_option("max_rows", 200)
pd.set_option("max_columns", 50)
```

```python
warnings.simplefilter("ignore")
```

```python
df = pd.read_pickle("../input/riiid-test-answer-prediction/split10/train_0.pickle")
```

```python
df = df[df["user_id"] % 100 == 0]
```

```python
df["tags"] = [str(x).split(" ") for x in df["tags"].values]
```

```python
tags_list = set()
```

```python
for xs in df["tags"].values:
    for x in xs:
        tags_list.add(x)
```

```python
tags_list
```

```python
result_dicts = []
for x in tqdm.tqdm(tags_list):
    df[f"tags_{x}"] = [1 if x in tags else 0 for tags in df["tags"].values]
    result = {}
    result["tags"] = x
    result["count"] = df[f"tags_{x}"].sum()
    result["targetencoding"] = df[df[f"tags_{x}"] == 1]["answered_correctly"].replace(-1, np.nan).mean()
    result_dicts.append(result)
```

```python
df_result = pd.DataFrame(result_dicts)
```
